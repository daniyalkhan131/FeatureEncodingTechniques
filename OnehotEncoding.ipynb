```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(df.iloc[:,0:4],df.iloc[:,-1],test_size=0.2,random_state=2)
X_train.head()
```

```python
X_train.sample(5)
```

```python
#we need to take out cols then apply encoding to fuel and owner make cols then
#-combine them
#we do this col. transformers
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
ohe = OneHotEncoder()
```

```python
X_train_new = ohe.fit_transform(X_train[['fuel','owner']]).toarray()
X_test_new = ohe.transform(X_test[['fuel','owner']]).toarray()
X_train_new.shape
```

```python
np.hstack((X_train[['brand','km_driven']].values,X_train_new))
```

```python
#but have to drop first col after encoding to remove multi.coll
ohe = OneHotEncoder(drop='first')
X_train_new = ohe.fit_transform(X_train[['fuel','owner']]).toarray()
X_test_new = ohe.transform(X_test[['fuel','owner']]).toarray()
print(X_train_new.shape)
np.hstack((X_train[['brand','km_driven']].values,X_train_new))
```

```python
#if we do sparse=False then we get numpy array dirct not sparse array
ohe = OneHotEncoder(drop='first',sparse=False,dtype=np.int32)
X_train_new = ohe.fit_transform(X_train[['fuel','owner']])
X_test_new = ohe.transform(X_test[['fuel','owner']])
print(X_train_new.shape)
np.hstack((X_train[['brand','km_driven']].values,X_train_new))
#we can also tell type of numpy array
```
